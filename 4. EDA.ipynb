```r
# Read in all files from the Data/Processed directory and create a list of data frames for each highway system
england_fnames <- list.files("Data/Processed/Highways_England/", pattern="*.csv", full.names=TRUE)
england_df_list <- lapply(england_fnames, read_csv)

portland_fnames <- list.files("Data/Processed/Portland/", pattern="*.csv", full.names=TRUE)
portland_df_list <- lapply(portland_fnames, read_csv)

utah_fnames <- list.files("Data/Processed/Utah/", pattern="*.csv", full.names=TRUE)
utah_df_list <- lapply(utah_fnames, read_csv)
```

```r
# Ensure that the rows in each data frame are in proper chronological order 
england_df_list <- lapply(england_df_list, function(x) x %>% arrange(timestamp))
portland_df_list <- lapply(portland_df_list, function(x) x %>% arrange(timestamp))
utah_df_list <- lapply(utah_df_list, function(x) x %>% arrange(timestamp))           
```

```r
# Create one list of all data frames
total_df_list <- append(append(england_df_list, portland_df_list), utah_df_list)
```

```r
# Randomly create a list of starting points from which we we sample each data frame

# Set the seed
set.seed(54321)

# Sample without replacement as to ensure the starting time for each time series is different
starting_points <- sample(1:((96*365)-(96*12*7)), length(total_df_list), replace=FALSE)

# Create the ending points by adding on 12 weeks to the starting points
ending_points <- starting_points + (96*7*12) - 1
```

```r
# Write the sample start/end points to a csv file for future use
points_df <- data.frame(start=starting_points, end=ending_points)
write.csv(points_df, 'start_end_points.csv', row.names=FALSE)
```

```r
# Grab 1 time series from each highway system for detailed EDA
# Do this by randomly sampling an index from 1 to the number of time series from each system
set.seed(54321)
eng_eda <- sample(1:length(england_df_list), 1)
por_eda <- length(england_df_list) + sample(1:length(portland_df_list), 1)
uta_eda <- length(england_df_list) + length(portland_df_list) + sample(1:length(utah_df_list), 1)
```

```r
# Grab the engalnd sample tim series using the above randomly generated index
# Get the highway name and sensor name from the data frame to be used for plotting, etc.
england_sample <- total_df_list[[eng_eda]]
england_highway_name <- str_split(england_sample$site_name[1], '/')[[1]][1]
england_sensor_name <- str_split(england_sample$site_name[1], '/')[[1]][2]
```

```r
# Grab the portland sample tim series using the above randomly generated index
# Get the highway name and sensor name from the data frame to be used for plotting, etc.
portland_sample <- total_df_list[[por_eda]]
portland_highway_name <- str_split(portland_sample$site_name[1], '/')[[1]][1]
portland_sensor_name <- str_split(portland_sample$site_name[1], '/')[[1]][2]
```

```r
# Grab the utah sample tim series using the above randomly generated index
# Get the highway name and sensor name from the data frame to be used for plotting, etc.
utah_sample <- total_df_list[[uta_eda]]
utah_highway_name <- str_split(utah_sample$site_name[1], '/')[[1]][1]
utah_sensor_name <- str_split(utah_sample$site_name[1], '/')[[1]][2]
```

```r
# Add a new column to each sample data frame to denote whether each row should be part of the training set, 
# validation set, or test set - the first 8 weeks are designated as train, the next 2 as val, 
# and the final 2 as test. 

england_sample <- england_sample[starting_points[eng_eda]:ending_points[eng_eda], ] %>% 
    mutate(rn = row_number()) %>%
    mutate(train_val_test = ifelse(rn <= (96*7*8), "train", ifelse(rn <= (96*7*10), "val", "test")))

portland_sample <- portland_sample[starting_points[por_eda]:ending_points[por_eda], ] %>%
    mutate(rn = row_number()) %>%
    mutate(train_val_test = ifelse(rn <= (96*7*8), "train", ifelse(rn <= (96*7*10), "val", "test")))

utah_sample <- utah_sample[starting_points[uta_eda]:ending_points[uta_eda], ] %>%
    mutate(rn = row_number()) %>%
    mutate(train_val_test = ifelse(rn <= (96*7*8), "train", ifelse(rn <= (96*7*10), "val", "test")))
```

# Example Exploratory Plots

## England

```r
# Create a basic density plot of the 15-minute traffic volume for the sample England sensor
eng_dens_plot <- ggplot(england_sample %>% filter(train_val_test == "train")) + 
    geom_density(aes(x=total_volume)) + 
    labs(y="Traffic Volume Density", x = "15-minute Traffic Volume") + 
    ggtitle(str_glue("Density Plot of 15-Minute Traffic Volume for the {england_highway_name} at Sensor {england_sensor_name}"))
```

```r
# Show the plot
eng_dens_plot
```

```r
# Create a basic time series plot of the first 3 weeks of the training portion of the England sample data
eng_ts_plot <- ggplot(england_sample %>% filter(rn <= (96*7*3))) + 
    geom_line(aes(x=timestamp, y=total_volume)) + 
    labs(y="15-minute Traffic Volume", x = "Timestamp") + 
    ggtitle(str_glue("3-Week Sample of 15-Minute Traffic Volume for the {england_highway_name} at Sensor {england_sensor_name}"))
```

```r
eng_ts_plot
```

```r
# Create a seasonal plot of the train portion of the England sample data which showcases the daily
# and day of week (weekly) seasonal patterns in the data
eng_daily_plot <- ggplot(england_sample %>% filter(train_val_test == "train")) + 
    geom_point(aes(x=interval_of_day, y=total_volume, col=as.factor(day_of_week)), alpha=0.5) + 
    labs(y="15-minute Traffic Volume", x = "Interval of Day", col="Day of Week") + 
    ggtitle(str_glue("Daily Pattern of 15-Minute Traffic Volume for the {england_highway_name} at Sensor {england_sensor_name}"))
```

```r
eng_daily_plot
```

```r
# Compute the ACF and intervals of the train portion of the England sample
eng_acf <- acf((england_sample %>% filter(train_val_test == "train"))$total_volume, lag=96*8, plot=FALSE)
eng_acf_ci_lo <- 1.96/(nrow(england_sample %>% filter(train_val_test == "train"))^0.5)
eng_acf_ci_hi <- -1*1.96/(nrow(england_sample %>% filter(train_val_test == "train"))^0.5)
```

```r
# Create an ACF plot
eng_acf_plot <- ggplot() + 
    geom_hline(aes(yintercept = 0)) +
    geom_segment(aes(x=eng_acf$lag, xend = eng_acf$lag, y=eng_acf$acf, yend = 0)) +
    geom_hline(aes(yintercept = eng_acf_ci_lo), col="blue", linetype=2) +
    geom_hline(aes(yintercept = eng_acf_ci_hi), col="blue", linetype=2) +
    labs(y="ACF", x="Lag") + 
    ggtitle(str_glue("ACF of 15-Minute Traffic Volume for the {england_highway_name} at Sensor {england_sensor_name}"))
```

```r
eng_acf_plot
```

```r
# Save all plots
ggsave("Results/EDA/eng_volume_dens.png", eng_dens_plot, width=10, height=5)
ggsave("Results/EDA/eng_ts_sample.png", eng_ts_plot, width=10, height=5)
ggsave("Results/EDA/eng_daily_pattern.png", eng_daily_plot, width=10, height=5)
ggsave("Results/EDA/eng_acf.png", eng_acf_plot, width=10, height=5)
```

# Comparisons Across Sensors
In this section, we examine all sensors together instead of looking at the details of a few sample sensors

```r
# Create a list of all data frames
total_df_list <- lapply(total_df_list, function(x) x %>% mutate(rn = row_number()))
```

```r
# Subset the data frames in the list to only the 12-week sample periods
total_df_list_samples <- lapply(1:length(total_df_list), 
                                function(x) total_df_list[[x]] %>% 
                                    filter(rn >= starting_points[x]) %>%
                                    filter(rn <= ending_points[x]))
```

```r
# Add the train, val, test column to denote what each observation in each df belongs to
total_df_list_samples <- lapply(total_df_list_samples, 
                                function(x) x %>% mutate(rn = row_number()) %>%
                                    mutate(train_val_test = ifelse(rn <= (96*7*8), 
                                                                   "train", 
                                                                   ifelse(rn <= (96*7*10), 
                                                                          "val", 
                                                                          "test")))
                                
                               )
```

```r
# Sanity check lenghts of each sample
lapply(total_df_list_samples, function(x) x %>% group_by(train_val_test) %>% summarise(n=n()))
```

```r
# Create a new list with only the train portstion of each time series
total_df_list_train <- lapply(total_df_list_samples, function(x) x %>% filter(train_val_test == "train"))
```

```r
# Copmute the mean and standard deviation of 15-minute volume for each time series
means <- unlist(lapply(total_df_list_train, function(x) mean(x$total_volume)))
sds <- unlist(lapply(total_df_list_train, function(x) sd(x$total_volume)))  
cols <- c(rep("England", length(england_df_list)), 
          rep("Portland", length(portland_df_list)),
          rep("Utah", length(utah_df_list)))                   
```

```r
# Create a box plot of the means with the highway system on the x axis
mean_box_plot <- ggplot() + 
    geom_violin(aes(x=cols, y=means)) + 
    geom_boxplot(aes(x=cols, y=means), width=0.1) + 
    labs(x="Highway System", y="Mean 15-minute Traffic Volume") + 
    ggtitle("Distribution of Mean 15-minute Traffic Volume per Sensor")
```
